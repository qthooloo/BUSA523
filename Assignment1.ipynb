```python
import random
randomNumber = random.randint(1,20)

print("A random number has been generated between 1-20. You have 5 attempts to guess it")
for number in range(5):
    test = input("Input your guess, then press enter")
    try:
        test = int(test)
    except ValueError:
        print("Input must be an integer, try again.")
        continue
    if test == randomNumber:
        win = True
        break
    else:
        print("Incorrect, try again")
        win = False
        continue        
if win == True:
    print("You win!")
else:
    print("You lose. Better luck next time.")
```

```python
import random

def rules():
    print("-----")
    print("Rock")
    print("Wins against: Lizard, Scissors")
    print("Loses to: Spock, Paper")
    print("-----")
    print("Paper")
    print("Wins against: Rock, Spock")
    print("Loses to: Scissors, Lizard")
    print("-----")
    print("Scissors")
    print("Wins against: Paper, Lizard")
    print("Loses to: Rock, Spock")
    print("-----")
    print("Lizard")
    print("Wins against: Spock, Paper")
    print("Loses to: Rock, Scissors")
    print("-----")
    print("Spock")
    print("Wins against: Rock, Spock")
    print("Loses to: Scissors, Lizard")

def playRound():
    oppMove = ""
    result = ""
    randomNumber = random.randint(1,5)
    if randomNumber == 1:
        oppMove = "Rock"
    elif randomNumber == 2:
        oppMove = "Paper"
    elif randomNumber == 3:
        oppMove = "Scissors"
    elif randomNumber == 4:
        oppMove = "Lizard"
    elif randomNumber == 5:
        oppMove = "Spock"
    ownMove = ""
    print("[1] - Rock")
    print("[2] - Paper")
    print("[3] - Scissors")
    print("[4] - Lizard")
    print("[5] - Spock")
    try:
        ownMove = int(input("Enter the number for your move and press enter: "))
        if int(ownMove) not in (1,2,3,4,5):
            result = "invalid"
    except ValueError:
        result = "invalid"

    if ownMove == randomNumber:
        result = "draw"
    elif ownMove == 1:
        if randomNumber == 3 or randomNumber == 4:
            result = "win"
        elif randomNumber == 2 or randomNumber == 5:
            result = "lose"
    elif ownMove == 2:
        if randomNumber == 1 or randomNumber == 5:
            result = "win"
        elif randomNumber == 3 or randomNumber == 4:
            result = "lose"
    elif ownMove == 3:
        if randomNumber == 2 or randomNumber == 4:
            result = "win"
        elif randomNumber == 1 or randomNumber == 5:
            result = "lose"
    elif ownMove == 4:
        if randomNumber == 2 or randomNumber == 5:
            result = "win"
        elif randomNumber == 1 or randomNumber == 3:
            result = "lose"
    elif ownMove == 5:
        if randomNumber == 1 or randomNumber == 3:
            result = "win"
        elif randomNumber == 2 or randomNumber == 4:
            result = "lose"
    return (result,oppMove)
option = ""
highScore = 0
score = 0

while option != '3':
    if option == '1':
        print("\033[H\033[J", end="")
        result = playRound()
        if result[0] == "win":
            score += 1
            if score > highScore:
                highScore = score
            print("\033[H\033[J", end="")
            print("Your opponent countered with " + result[1])
            print("You win!")
            print("High Score: " + str(highScore))
            print("Current Score: " + str(score))
        elif result[0] == "lose":
            print("\033[H\033[J", end="")
            print("Your opponent countered with " + result[1])
            print("Game Over!")
            print("High Score: " + str(highScore))
            print("Current Score: " + str(score))
            score = 0          
        elif result[0] == "draw":
            print("\033[H\033[J", end="")
            print("Your opponent countered with " + result[1])
            print("It's a draw.")
            print("High Score: " + str(highScore))
            print("Current Score: " + str(score))
        elif result[0] == "invalid":
            print("Invalid move, try again.")
        option = ""
        input("Press 'Enter' to continue..")
        continue
    elif option == '2':
        print("\033[H\033[J", end="")
        rules()
        option = ""
        input("Press 'Enter' to continue..")
        continue
    else:
        print("\033[H\033[J", end="")
        print("Rock, Paper, Scissors, Lizard, Spock.")
        print("High Score: " + str(highScore))
        print("[1] Start / Next Round")
        print("[2] Rules")
        print("[3] Exit")
        option = input("Select an option and press enter:")
```
