```python
import random
import numpy

results = []

for num in numpy.arange(0,10000):
    coinFlipResult = random.randint(0,1)
    if coinFlipResult == 0:
        coinFlipResult = 'Heads'
    else:
        coinFlipResult = 'Tails'
    results.append(coinFlipResult)

streakCount = 0
potStreakNum = 0
highStreak = 0

for index,result in enumerate(results):
    if index == 0 or result != results[index-1] or (index + 1 == len(results)):
        if potStreakNum >= 5:
            streakCount += 1
            if streakCount >= highStreak:
                highStreak = potStreakNum
                highStreakType = results[index-1]
        potStreakNum = 1
        continue
    elif result == results[index-1]:
        potStreakNum += 1
        continue
print("Streaks of 5 or greater: " + str(streakCount))
print("Highest Streak: " + highStreakType + " - " + str(highStreak))
```

```python
inventory = {}
for item in ("swords","diamonds","gold coins","rope","potions"):
    itemQuantity = ""
    while itemQuantity == "":
        try:
            itemQuantity = int(input(("Enter your desired number of " + item)))
        except ValueError:
            print("Input must be an integer. Please try again.")
            itemQuantity = ""
    inventory[item] = itemQuantity
        
print("Inventory:")
for item in inventory:
    print(str(inventory.get(item)), item)
```

```python
import random
import time

class Player:
    def __init__(self, id):
        self.id = "Player " + str(id)
    
    score = 0

    def rollDice(self):
            print(self.id + " rolls the dice...")
            result = random.randint(1,6)
            time.sleep(1.5)
            print(self.id + " rolled a " + str(result))
            self.score += result

numPlayers = 3
numRounds = 10
numSides = 6
players = []
finalScores = {}

for num in range(1,numPlayers+1):
    players.append(Player(num))

for num in range(0,numRounds):
    print("====Round " + str(num+1) + "====")
    for p in players:
        p.rollDice()
        time.sleep(2)
    
for p in players:
    finalScores[p.id] = p.score

finalScores = sorted(finalScores.items(), key=lambda x: x[1], reverse=True)
print("====   Winner   ====")
print(finalScores[0][0])

print("====Final Scores====")
print(finalScores)
```
