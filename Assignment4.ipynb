```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime as dt

data = pd.read_csv("data.csv", header = 0, index_col = 0)
```

Histogram of Temperature
========================
---


```python
plt.hist(data["Temperature"])
```

Bar Chart of Unemployment
===
---


```python
unemployment = data["Unemployment"]
cleanedUnemployment = unemployment.drop_duplicates()
cleanedUnemployment.plot.bar()
```

Line Chart of Gas Prices
===
---

```python
fuelPrice = pd.DataFrame(data["Fuel_Price"])
fuelPrice.plot(rot=45)
```

Area Plot of Sales by Department
===
---

```python
salesData = pd.DataFrame({'Dept 1':data["dept_1 sales"], 'Dept 2':data["dept_2 sales"], 'Dept 3':data["dept_3 sales"]})

salesData.plot.area(rot=45)
```
