```python
class Student:
    
    def __init__(self, fname, lname, country, gpa, credits, major):
        self.fname = fname
        self.lname = lname
        self.country = country
        self.gpa = gpa
        self.credits = credits
        self.major = major
        # Set initial progress
        if self.gpa > 3.0 and self.credits > 20:
            self.progress = "Good"
        elif self.gpa > 2.5 or 10 < self.credits < 20:
            self.progress = "Average"
        elif self.gpa < 2.5:
            self.progress = "Bad"
        else:
            self.progress = "Undefined"
    
    def updateProgress(self):
        if self.gpa > 3.0 and self.credits > 20:
            self.progress = "Good"
        elif self.gpa > 2.5 or 10 < self.credits < 20:
            self.progress = "Average"
        elif self.gpa < 2.5:
            self.progress = "Bad"
        else:
            self.progress = "Undefined"
        return self.progress
            
goodStudent = Student("Good", "Student", "USA", 4.0, 26, "MBUSA")
averageStudent = Student("Average", "Student", "USA", 3.0, 15, "Accounting")
badStudent = Student("Bad", "Student", "USA", 1.7, 12, "Math")
newStudent = Student("New", "Student", "USA", 4.0, 6, "CompSci")

print(goodStudent.progress)
print(averageStudent.progress)
print(badStudent.progress)
print(newStudent.progress)
```

```python
def sortPosNeg(numbers):
    negative = []
    positive = []
    for x in numbers:
        if x > 0:
            positive.append(x)
        elif x < 0:
            negative.append(x)    
    print(negative, positive)

sample = [-6, 5, -3, -2, 1, 0, -8, 9, 3]
sortPosNeg(sample)
```

```python
def listPossibilities(playerScore):
    winPossibilities = []
    for x in range(12):
        for y in range(12):
            if playerScore < x + y <= 21:
                winPossibilities.append((x,y))
    print(winPossibilities)

listPossibilities(15)
```
